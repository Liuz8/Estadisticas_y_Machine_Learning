## Paso 1

```python
X = df.drop('churn',axis=1) # variable explicativa
y = df['churn'] # variable respuesta
```

```python
X
```

```python
y
```

## Paso 2

```python
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
```

```python
columnas = X.columns
```

```python
one_hot = make_column_transformer((OneHotEncoder(drop='if_binary'),
                                   ['pais', 'sexo_biologico', 'tiene_tarjeta_credito', 'miembro_activo']),
                                   remainder='passthrough',
                                   sparse_threshold=0,
                                   force_int_remainder_cols=False)
```

```python
X = one_hot.fit_transform(X)
```

```python
one_hot.get_feature_names_out(columnas)
```

```python
X
```

```python
pd.DataFrame(X,columns=one_hot.get_feature_names_out(columnas))
```

## Paso 3

```python
label_encoder = LabelEncoder()
```

```python
y = label_encoder.fit_transform(y)
```

## Paso 1

```python
from sklearn.model_selection import train_test_split
```

```python
X_test,X_train,y_test,y_train = train_test_split(X,y,stratify=y, random_state=5)
```

## Paso 2

```python
from sklearn.dummy import DummyClassifier
```

```python
dummy = DummyClassifier()
dummy.fit(X_train,y_train)

dummy.score(X_test,y_test)
```

## Paso 3

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
modelo_arbol = DecisionTreeClassifier(max_depth=4,random_state=5)
modelo_arbol.fit(X_train, y_train)
```

```python
modelo_arbol.score(X_test,y_test)
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
```

```python
columnas_transformadas = one_hot.get_feature_names_out()
```

```python
columnas_transformadas
```

```python
plt.figure(figsize=(80,25))
plot_tree(modelo_arbol,filled=True,class_names=['no','si'],fontsize=6,feature_names=columnas_transformadas);
```
